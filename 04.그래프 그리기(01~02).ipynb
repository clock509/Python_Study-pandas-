```python
#앤스콤 데이터 집합 불러오기.
import seaborn as sns

anscombe = sns.load_dataset("anscombe")  #.load_dataset("데이터명")
print(anscombe)
print(type(anscombe))

#matplotlib 라이브러리로 그래프 그리기
%matplotlib notebook
import matplotlib.pyplot as plt

#dataset 열에서 데이터값이 I인 것만 추출
dataset_1 = anscombe[anscombe['dataset'] == 'I']
plt.plot(dataset_1['x'], dataset_1['y'])
```

```python
#plot은 기본값이 선. 점으로 그리려면 'o'를 세번째 인자로 전달.
%matplotlib notebook
import matplotlib.pyplot as plt

plt.plot(dataset_1['x'], dataset_1['y'], 'o')
```

#matplotlib 라이브러리 사용하기

1. 전체 그래프가 위치할 기본 틀을 만든다.
2. 그래프를 그릴 격자를 만든다.
3. 격자에 그래프를 하나씨 추가한다. 왼쪽에서 오른쪽 방향으로 하나씩 추가된다.
4. 격자의 첫 번째 행이 꽉 차면 두 번째 행에 그래프를 그려 넣는다.

```python
#1. 데이터를 준비한다(anscombe 데이터프레임을 dataset 열의 값대로 불린 추출)
%matplotlib notebook
import matplotlib.pyplot as plt
import seaborn as sns

dataset1 = anscombe[anscombe['dataset']=='I']
dataset2 = anscombe[anscombe['dataset']=='II']
dataset3 = anscombe[anscombe['dataset']=='III']
dataset4 = anscombe[anscombe['dataset']=='IV']
print(type(dataset1))

#2. 그레프 격자를 그릴 기본 틀(=4개의 그래프를 담을 큰 캔버스) 만들기.
fig = plt.figure()

#2-1. add_subplot으로 기본 틀(fig)에 그래프 격자 그리기.
axes1 = fig.add_subplot(2, 2, 1)
axes2 = fig.add_subplot(2, 2, 2)
axes3 = fig.add_subplot(2, 2, 3)
axes4 = fig.add_subplot(2, 2, 4)

#2-2. plot 메서드에 데이터를 전달하고 그린다.
axes1.plot(dataset1['x'], dataset1['y'], 'o')
axes2.plot(dataset2['x'], dataset2['y'], 'o')
axes3.plot(dataset3['x'], dataset3['y'], 'o')
axes4.plot(dataset4['x'], dataset4['y'], 'o')

#2-3. 각각의 그래프 격자에 제목 추가(.set_title())
axes1.set_title("dataset1")
axes2.set_title("dataset2")
axes3.set_title("dataset3")
axes4.set_title("dataset4")

#2-4. 기본 틀에 제목 붙이기(.suptitle()) & 윗 그래프와 아래 그래프가 서로 겹치지 않게 레이아웃 조절(.tight_layout)
fig.suptitle("Anscombe Data")
fig.tight_layout()
print("\n", fig)
```

4-2. matplotlib 라이브러리 자유자재로 사용하기

```python
import seaborn as sns
%matplotlib notebook
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
print(tips.head())
print(type(tips))

#히스토그램: 열 데이터의 분포와 빈도를 살펴보는 용도.
##일변량 그래프: 변수를 하나만 사용해서 그린 그래프

#히스토그램을 그릴 준비
fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)

#히스토그램 생성(hist)
axes1.hist(tips['total_bill'], bins = 10)  #bins: x축 간격
axes1.set_title('Histogram of Total Bill') #그래프 제목
axes1.set_xlabel('Frequency') #x축 제목(일번량 그래프이므로, x축은 데이터 갯수만큼 생성될 것)
axes1.set_ylabel('Total Bill')

#산점도
##이변량 그래프: 변수 2개를 사용하여 만듬. x에 따른 y의 분포를 나타냄.

#산점도를 그릴 준비
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)

#산점도 생성(scatter)
axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

```python
#박스 그래프(Boxplot)
#이산형 변수와 연속형 변수를 모두 사용.

#박스플롯 그릴 준비
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1, 1, 1)

#boxplot 그리기
axes1.boxplot([tips[tips['sex'] == 'Female']['tip'],
             tips[tips['sex'] == 'Male']['tip']],
             labels = ['Female', 'Male'])

axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

```python
#다변량 그래프: 점의 크기 or 투명도를 조절하여 내보임.
#잎의 산점도 그래프에 성별을 새 변수로 추가해 보자.

#Male, Female같은 문자를 그래프에 나타낼 수 없으므로 0, 1로 범주화(or 치환)해야 한다.
def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1
#recode_sex가 반환한 0, 1을 데이터프레임에 추가하자(apply 메서드)
tips['sex_color'] = tips['sex'].apply(recode_sex)
print(tips.head())

#산점도에도 추가해보자.
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(
x = tips['total_bill'],   #x축 변수
y = tips['tip'],          #y축 변수
s = tips['size'] * 10,    #s = 점 사이즈
c = tips['sex_color'],    #c = 점 색상
alpha = 0.5)              #alpha = 투명도
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```
