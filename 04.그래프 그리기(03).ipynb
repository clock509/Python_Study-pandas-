04-3. seaborn 라이브러리 자유자재로 사용하기

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
anscombe = sns.load_dataset("anscombe")
```

```python
#subplot, displot: 히스토그램
#displot - kde  : 밀집도 
#displot - hist : 히스토그램
ax = plt.subplots() #subplot: 기본 틀
ax = sns.distplot(tips['total_bill']) #displot: 그리고자 하는 데이터 전달
                                      #displot: 히스토그램(bar shaped), 밀집도(curve shaped) 그래프
ax.set_title('Total Bill Histogram with Density Plot')
```

```python
#subplot, displot: 히스토그램
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], kde = False) #위 그래프에서 밀집도 그래프(곡선)를 제외하고 싶다면: kde = False  
ax.set_title('Total Bill Histogram with Density Plot')
ax.set_xlabel('Total Bill') #x축명
ax.set_ylabel('Frequency')  #y축명
```

```python
#밀집도 그래프만 나타내려면: hist = False
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist = False)
ax.set_title('Total Bill Density')
ax.set_xlabel('Total Bill') #x축명
ax.set_ylabel('Unit Probability')  #y축명
```

```python
#displot + rug(양탄자 그래프)
#실무 환경에서는 데이터 시각화를 더 효율적으로 하기 위해 여러 그래프를 한 번에 출력하는 경우가 많다.
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug = True)
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
```

```python
#count 그래프: histogram과 비슷하게 생김. count 그래프는 이산값을 넣어 나타낸 그래프. 즉 빈도를 나타냄.
ax = plt.subplots()
ax = sns.countplot('day', data = tips)
ax.set_title('Count of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

다양한 종류의 이변량 그래프 그리기: 산점도 그래프 by seaborn

```python
#1. regplot : 산점도 그래프
ax = plt.subplots()
ax = sns.regplot(data = tips, x = 'total_bill', y = 'tip')
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('tip')

ax = plt.subplots()
ax = sns.regplot(data = tips, x = 'total_bill', y = 'tip', fit_reg = False) #fit_reg = False ==> 회귀선 제거
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
#2. jointplot : 산점도와 히스토그램을 한 번에 그려줌

joint = sns.jointplot(data = tips, x = 'total_bill', y = 'tip')
joint.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip')
joint.fig.suptitle('Joint Plot of  Total Bill and Tip', fontsize = 18, y = 1.03)
```

```python
#3. hexbin: 육각 그래프(점 그래프의 단점 커버: 점이 겹쳐 보일 경우 점을 구분하기 어려움)
#jointplot을 그리되 kind인자를 hex로 지정
hexbin = sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex")
hexbin.set_axis_labels(xlabel= "Total Bill", ylabel = "TIp")
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize = 10, y = 1)
```

```python
#4. 이차원 밀집도 그리기
#kdeplot 메서드
ax = plt.subplots()
ax = sns.kdeplot(data = tips['total_bill'], data2 = tips['tip'], shade = True) #shade인자 : 그래프에 음영 효과
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
#5. 바 그래프
#지정한 변수의 평균을 계싼하여 그린다.
ax = plt.subplots()
ax = sns.barplot(data = tips, x = 'time', y = 'total_bill') #시간에 따른 팁 금액
ax = sns.barplot(data = tips, x = 'day', y = 'total_bill')  #요일에 따른 팁 금액
ax.set_title('Barplot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

```python
#6. 박스 그래프
#최솟값, 1분위수, 중간값, 3분위수, 최댓값, 이상치 등 통계량을 한 번에 표현
ax = plt.subplots()
ax = sns.boxplot(data = tips, x = 'time', y = 'total_bill')
ax.set_title('Boxplot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
```

```python
#박스그래프의 단점: 분산이 모호하게 표현됨
#박스그래프에 커널 밀도를 추정한 바이올린 그래프를 사용!
#boxplot 대신 violinplot 명령어 사용
ax = plt.subplots()
ax = sns.violinplot(data = tips, x = 'time', y = 'total_bill')
ax.set_title('Violinplot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total BIll')
```

```python
#7. 관계 그래프: 위 그래프들을 종합
#pairplot 메서드
fig = sns.pairplot(tips)
#단점: 대각선을 기준으로 각 절반은 서로 같다.
#(1, 0) = (0, 1)
#(2, 0) = (0, 2)
```

```python
#map_upper : 대각선을 기준으로 위쪽에 그릴 그래프 지정
#map_lower : 대각선을 기준으로 아래쪽에 그릴 그래프 지정
#map_diag : 대각선을 중심으로 그래프를 그림
#대각선 아래는 이차원 밀집도, 위쪽은 산점도, 대각선에는 히스토그램과 밀집도
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot)
pair_grid = pair_grid.map_diag(sns.distplot, rug = True)
plt.show()
```

다변량 그래프 그리기 by seaborn

```python
#X = time, Y = total_bill의 분산을 남여 구분하여 보여주기
ax = plt.subplots()
ax = sns.violinplot(data = tips, x = 'time', y = 'total_bill', hue = 'sex', split = True) #hue: 색상에 사용할 열 이름
```

```python
#산점도, 관계 그래프에 색상 추가하기
scatter = sns.lmplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex', fit_reg = False)
fig = sns.pairplot(tips, hue = 'sex')
```

```python
#산점도 그래프: 크기와 모양 조절하기(크기: 딕셔너리 형태로 전달, 모양: 리스트 형태로 전달)
#방법: scatter_kws에 '딕셔너리' 형태로 인잣값을 전달한다.
scatter = sns.lmplot(data = tips, 
                     x = 'total_bill', 
                     y = 'tip', 
                     fit_reg = False, hue = 'sex', 
                     scatter_kws = {'s': tips['size']*5})  #'s':tips['size']를 딕셔너리에 담아 전달
```

```python
#점 모양 바꾸기: markers 인잣값에 표현하고자 하는 기호를 '리스트'로 담아 전달.
scatter = sns.lmplot(data = tips,
                    x = 'total_bill',
                    y = 'tip',
                    fit_reg = False,
                    hue = 'sex',
                    markers = ['o', 'x'],
                    scatter_kws = {'s': tips['size']* 10})
```

```python
#lmplot 메서드: 4개의 데이터 그룹에 대한 그래프 그리기
#앤스콤 4분할 그래프 그리기
anscombe_plot = sns.lmplot(x = 'x', y = 'y', data = anscombe, fit_reg = False) #fit_reg ==> 회귀선 유무 지정
#이렇게만 하면 4개의 데이터 그룹이 한꺼번에 산점도 그래프로 그려진다. 그룹별로 나눠서 그려보자.
```

```python
#위의 anscombe그래프를 나눠서 그리려면? ==> col, col_warp 인자를 설정한다.
#col_wrap = 그래프를 그릴 열의 최댓값
#col = 데이터 그룹을 구분할 열(dataset) 지정
anscombe_plot = sns.lmplot(x = 'x', y = 'y', data = anscombe, fit_reg = False, col = 'dataset', col_wrap = 2)
```

FacetGrid 클래스로 그룹별 그래프 그리기

```python
#FacetGrid에 데이터프레임과 그룹을 구분할 열 이름을 전달하여 객체를 생성
#객체(facet)의 map 메서드에 그래프의 종류(displot)와 x축으로 지정할 열 이름을 순서대로 전달하면 
#히스토그램, 밀집도, 양탄자(rug) 그래프 그릴 수 있음.

facet = sns.FacetGrid(tips, col = 'time')
facet.map(sns.distplot, 'total_bill', rug = True)
```

```python
#FacetGrid로 다변량 그래프 그리기
facet = sns.FacetGrid(tips, col = 'day', hue = 'sex')
facet = facet.map(plt.scatter, 'total_bill', 'tip')  #scatter: 산점도
facet = facet.add_legend()
```

```python
#각각 2개의 값을 가진 범주형 변수 2개의 그래프 그리기 ==> 총 4개의 그래프 나옴(A1, B1), (A1, B2), (A2, B1), (A2, B2)
facet = sns.FacetGrid(tips, col = 'time', row = 'smoker', hue = 'sex')
facet.map(plt.scatter, 'total_bill', 'tip')
```
